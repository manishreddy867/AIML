## 2. What is Regression?

- Predicting a **label (dependent variable)** from **features (independent variables)**
- Used for: **forecasting**, **time-series**, finding **cause-effect** relationships
- A curve/line is fit to data so the distance between actual points and the line is **minimum**

## 3. Linear Regression – Core Idea

**Equation:**
$$Y = a + bX + e$$

- `a` → intercept (where line crosses Y-axis)
- `b` → slope (how much Y changes per unit of X)
- `e` → error term

**Goal:** Find the best values of `a` and `b` to predict Y accurately.

## 5. Simple Linear Regression

- Uses **one feature** to predict the target
$$y = \beta_0 + \beta_1 x$$

- `β₀` = intercept, `β₁` = coefficient (slope)
- Coefficients are learned using the **Least Squares method**

**Least Squares = minimize the sum of squared residuals (errors)**

Residual = actual Y − predicted Y

Best-fit line = line where $D_1^2 + D_2^2 + ... + D_N^2$ is minimum

## 6. Assumptions of Linear Regression

1. Model is **linear** in coefficients
2. **Mean of residuals = 0**
3. Error terms are **not correlated** with each other
4. Independent variables are **not correlated** with error terms (exogeneity)
5. Error terms have **constant variance** (homoscedasticity)
6. **No multicollinearity** between independent variables
7. Error terms are **normally distributed**

## 7. How the Model Learns – Gradient Descent

- The best `m` (slope) and `b` (intercept) are found by **minimizing the residual**
- Minimization is done via **partial derivatives** set to 0
- Since residual depends on both `m` and `b`, the graph becomes a **paraboloid**
- **Gradient Descent** is used to find the lowest point (minima)
- **Learning rate (α):** Controls how fast we descend. Recommended range: `0.00001` to `10`
- Smaller learning rate = more stable but slower

## 8. Model Evaluation Metrics

### R² (R-Squared)
- Measures **proportion of variance explained** by the model
- Range: 0 to 1 (closer to 1 = better)
- R² = 0.75 → model explains 75% of data
- R² < 0 → model is worse than a basic average line

$$R^2 = 1 - \frac{RSS}{TSS}$$

- **RSS** = Residual Sum of Squares (model error)
- **TSS** = Total Sum of Squares (error from using mean as predictor)

### Adjusted R²
- R² always increases when more features are added, even useless ones
- Adjusted R² **penalizes** adding irrelevant features
- Always ≤ R²
- Use this over plain R² when comparing models with different feature counts

## 9. Confidence Intervals & Hypothesis Testing

### Confidence Interval
- A 95% confidence interval means: if we sampled 100 times, ~95 intervals would contain the true coefficient
- We only have one sample, so uncertainty always exists

### p-value
- Probability that the coefficient is actually 0 (i.e., no effect)
- **p-value < 0.05** → reject null hypothesis → feature IS related to target
- **p-value ≥ 0.05** → fail to reject null → feature may NOT be significant
- If 95% CI includes 0 → p-value > 0.05 (feature is not significant)
- If 95% CI does NOT include 0 → p-value < 0.05 (feature is significant)

## 10. Multiple Linear Regression

- Uses **multiple features** to predict the target
$$y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n$$

**Example (Advertising dataset):**
$$y = \beta_0 + \beta_1 \cdot TV + \beta_2 \cdot Radio + \beta_3 \cdot Newspaper$$

**Findings:**
- TV and Radio → positively related to sales ✅
- Newspaper → negative or near-zero coefficient ❌
- R² improved with multiple features (0.897 vs single feature)

**Interpreting coefficients:**
- If Newspaper coefficient is **negative** → spending more on newspaper actually hurts sales (or has no real impact)

## 11. Feature Selection

**Simple approach:**
- Keep features with small p-values
- Check if R² improves when adding new features

**Drawbacks:**
- R² always increases even with random/noisy features
- 100 random features → ~5 will still appear significant by chance (false positives)

**Better approach: Cross-Validation**
- More reliable for selecting features
- Estimates how well model generalizes to new data
- Available in scikit-learn

## 12. Handling Categorical Variables

### Two Categories (e.g., small / large)
- Create a **dummy variable**: `small = 0`, `large = 1`
- Use this single column as a feature

### More than Two Categories (e.g., rural / suburban / urban)
- Cannot assign numbers directly (0, 1, 2) — implies false ordering
- Use **one-hot encoding** via `pd.get_dummies()`
- Drop one column to avoid the **Dummy Variable Trap**
  - Rule: N categories → use only N-1 dummy columns
  - The dropped category is captured when all others = 0

**Ordinal categories** (worst → bad → good → best) **can** be encoded as 1, 2, 3, 4 using one column

## 13. Multicollinearity

**What it is:** When independent variables are correlated with each other

**Why it's a problem:**
- Violates the assumption that features are independent
- Leads to redundancy and can cause overfitting
- Coefficients become unreliable

**Detection:**
- Correlation matrix: correlation > 0.9 or < -0.9 → problem
- **VIF (Variance Inflation Factor):**
  $$VIF = \frac{1}{1 - R^2}$$
  - VIF > 10 → severe multicollinearity
  - VIF > 5 → needs attention

**Remedies:**
- Ignore if correlation is mild
- Drop one of the correlated variables
- Combine correlated variables (e.g., age + experience → seniority score)
- Use **PCA** (Principal Component Analysis)

## 14. Bias-Variance Tradeoff

- **Bias** = how far predictions are from true values (model too simple → underfitting)
- **Variance** = how much predictions change with different training data (model too complex → overfitting)

$$\text{Total Error} = Bias^2 + Variance + \text{Irreducible Error}$$

- As bias decreases → variance increases (and vice versa)
- Goal: find a **sweet spot** with low bias AND low variance

**Linear Regression:** High bias / Low variance model
- The best-fit line stays roughly the same even with repeated sampling

**Fix:**
- Reduce bias → train with more data
- Reduce variance → use **Regularization**

## 15. Regularization

**Why?** Prevents overfitting by penalizing large coefficients

### L1 – LASSO Regression
$$\text{Loss} = RSS + \lambda \sum |\beta_j|$$
- Penalizes by **sum of absolute values** of coefficients
- Can reduce some coefficients **exactly to 0** → performs feature selection
- Good when only a few features are truly useful

### L2 – Ridge Regression
$$\text{Loss} = RSS + \lambda \sum \beta_j^2$$
- Penalizes by **sum of squared values** of coefficients
- Shrinks coefficients **close to 0** but never exactly 0
- Good default; works well when many features contribute a little

### Elastic Net
- Mix of Ridge and LASSO
- Controlled by mixing parameter α (0 = Ridge, 1 = LASSO)
- Best choice when features are many or correlated

### Choosing λ
- Done using **cross-validation**
- Higher λ → stronger regularization (more penalty)
- Too high → model starts underfitting

**When to use what:**
| Situation | Use |
|---|---|
| Default | Ridge |
| Only a few features matter | LASSO |
| Features are correlated or many | Elastic Net |
| Never | Plain OLS (no regularization) |

## 16. Polynomial Regression

**When to use:** When the relationship between X and Y is **not linear** (curved)

$$y = a + bx + cx^2 + ... + nx^n$$

- `Degree 1` → straight line (simple linear regression)
- `Degree 2` → quadratic (U-shaped curve)
- `Degree 4` → more complex curve

**Still uses `LinearRegression` in sklearn — why?**
- "Linear" refers to linearity in the **coefficients** (a, b, c...), NOT in x
- We just transform x → x², x³... first, then apply linear regression

**Implementation:**
- Use `PolynomialFeatures(degree=n)` from sklearn to create the polynomial features
- Then fit a normal `LinearRegression()` on the transformed features

**Dataset used:** Position_Salaries.csv — predicting salary from job level
- Linear model → poor fit
- Degree 2 → still not great
- Degree 4 → best fit ✅
