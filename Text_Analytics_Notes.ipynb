## 1. What is Text Analytics?

Text analytics is the process of extracting meaningful information from unstructured text data. In today's world, most companies collect massive amounts of text data — customer reviews, emails, social media posts, search queries, and surveys. All of this is raw and unstructured, meaning a machine cannot directly understand it without processing.

A very common real-world example is Google. When you search for a product on Google, you immediately start seeing ads for that product on other websites. This happens because Google analyzes your search text, identifies your intent, and passes that information to advertisers. That's text analytics in action.

Other common use cases include:
- **Sentiment Analysis** — analyzing customer reviews to detect if they are positive or negative
- **Product Recommendation** — recommending products based on search and purchase history
- **Spam Detection** — classifying emails as spam or not spam
- **Chatbots and Virtual Assistants** — understanding and responding to human queries

## 2. What is NLP (Natural Language Processing)?

NLP is a branch of Artificial Intelligence that enables machines to read, understand, and interpret human language. Human language is extremely complex — it has sarcasm, idioms, multiple meanings for the same word, different dialects, and grammatical rules that change by context. Teaching a computer to understand all of this is the core challenge of NLP.

The ultimate goal of NLP is not just to read text but to derive meaningful conclusions from it — like understanding the sentiment behind a sentence or extracting important entities like names and places.

**Famous examples of NLP applications:**
- **Siri** (Apple) — understands voice commands and responds intelligently

## 3. Tokenization

Before any NLP task, raw text needs to be broken down into smaller, manageable pieces. This process is called **Tokenization**. Think of it like cutting a large cake into slices — you can't eat the whole cake at once, so you cut it into pieces first. Similarly, a machine can't process an entire paragraph at once, so we break it down.

There are two types:
- **Sentence Tokenization** — splits a paragraph into individual sentences
- **Word Tokenization** — splits sentences into individual words (and punctuation marks)

**Example — Input paragraph:**
> "India (Hindi: Bhārat), officially the Republic of India, is a country in South Asia. It is the seventh-largest country by area..."

**Sentence Tokenization output:**
```
['India (Hindi: Bhārat), officially the Republic of India, is a country in South Asia.',
 'It is the seventh-largest country by area, the second-most populous country...',
 'Bounded by the Indian Ocean on the south...',
 'In the Indian Ocean, India is in the vicinity of Sri Lanka...']
```

**Word Tokenization output:**
```
['India', '(', 'Hindi', ':', 'Bhārat', ')', ',', 'officially', 'the', 'Republic', ...]
```

Notice that punctuation marks like `(`, `,`, `.` are also treated as separate tokens. This is important for later cleaning steps.

## 4. POS Tags (Parts of Speech Tagging)

Once a sentence is tokenized into words, the next step is to understand what role each word plays in the sentence. This is called **POS Tagging**. Every word in a sentence is a specific part of speech — noun, verb, adjective, adverb, etc. — and POS tagging assigns that label to each word automatically.

In traditional grammar, there are 9 parts of speech. But in NLP, these are broken down into much more specific tags. For example, nouns alone have 4 tags:
- `NN` → singular noun (e.g., table)
- `NNS` → plural noun (e.g., tables)
- `NNP` → proper noun singular (e.g., India)
- `NNPS` → proper noun plural (e.g., Indians)

Similarly, verbs, adjectives, and adverbs all have multiple subtags.

**Why is POS tagging useful?**
It helps the machine understand the grammatical structure of a sentence. This is critical for tasks like Named Entity Recognition, Parsing, and translation.

**Example:**
```python
data = 'We will see an example of POS tagging.'
pos = nltk.pos_tag(nltk.word_tokenize(data))
# Output: [('We', 'PRP'), ('will', 'MD'), ('see', 'VB'), ('an', 'DT'), ('example', 'NN'), ...]
```

## 5. Chunking

After POS tagging, sometimes we want to group multiple related words together under a single user-defined label. This is called **Chunking** (also known as shallow parsing). Instead of looking at each word individually, chunking lets us define rules to group words based on their POS tags.

For example, if you want to group all noun phrases together (a proper noun followed by a common noun), you can define that rule and chunking will apply it across the entire text automatically.

**Why chunking?** It helps structure unstructured text. Instead of just knowing that "Republic" is an NNP and "India" is an NNP, chunking lets you say — these two together form a meaningful noun phrase.

**Example:**
```python
my_node = "MN: {<NNP>*<NN>}"   # group any NNP followed by NN under tag 'MN'
chunk = nltk.RegexpParser(my_node)
result = chunk.parse(pos)
result.draw()   # visualizes the chunk tree graphically
```

The output is a tree structure where the custom tag `MN` groups together all proper nouns and nouns that appear together. This makes the data more organized and easier to extract meaning from.

## 6. Stop Words

In any piece of text, some words appear very frequently but carry almost no useful information on their own. Words like **'a', 'an', 'the', 'is', 'at', 'of', 'in'** appear in virtually every sentence, regardless of the topic. These are called **Stop Words**.

If we keep stop words in our dataset, they would dominate the word frequency counts and mislead the model. For example, if we're analyzing customer reviews, the word 'the' would appear thousands of times, but it tells us nothing about whether the review is positive or negative. Removing stop words reduces noise and makes the remaining words more meaningful.

NLTK provides a built-in list of stop words for many languages. We can also add custom words to this list based on our specific problem.

Similarly, **punctuation marks** like `.`, `,`, `!`, `?` are also removed during preprocessing since they add no semantic value.

```
['India', 'Hindi', 'Bhārat', 'officially', 'Republic', 'India', 'country', 'South', 'Asia', 'seventh-largest', ...]
```

## 7. Stemming

In everyday language, the same base word appears in many different forms depending on grammar — for example, **run, runs, running, ran** all refer to the same concept. If we treat these as 4 different words, the model gets confused. **Stemming** is the process of reducing each word to its base or root form by chopping off suffixes and prefixes.

The key point is that stemming does **not** care about whether the resulting stem is a real word or not. It just applies a set of rules to cut the word.

**Examples:**
- `computation` → `comput`
- `computer` → `comput`
- `hobbies` → `hobbi`

Notice that `comput` and `hobbi` are not real words — but that's okay for stemming. As long as both `computation` and `computer` map to the same stem, the model treats them as the same concept.

**Three stemmers available in NLTK:**

| Stemmer | Speed | Complexity | Notes |
|---|---|---|---|
| **Porter** | Moderate | Low | Oldest and simplest algorithm |
| **Lancaster** | Fast | High | More aggressive cuts, may over-stem |
| **Snowball (Porter2)** | Moderate | Medium | Updated version of Porter, currently most popular, supports multiple languages |

**Code:**
```python
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer

porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer('english')

print(porter.stem('running'))    # → run
print(lancaster.stem('running')) # → run
print(snowball.stem('running'))  # → run
```

Note: `porter.stem('ran')` → `ran` (not `run`). Stemming does not handle irregular verbs well.

## 8. Lemmatization

Lemmatization does the same job as stemming — reducing a word to its base form — but with one important difference: **it always returns a real, valid word from the dictionary**. The base form returned is called the **Lemma**.

For example:
- `running` → `run` (lemma)
- `better` → `good` (lemma — it understands comparative forms)
- `ran` → `run` (lemma — it handles irregular verbs if POS is provided)

**The catch — context matters.** Without telling the lemmatizer what part of speech the word is, it may return the wrong lemma. For example, `running` without POS context gives `running` back because it treats it as a noun. If you pass `pos='v'` (verb), it correctly gives `run`.

**Code:**
```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()

print(lemma.lemmatize('running'))        # → running (treated as noun, no context)
print(lemma.lemmatize('running', pos='v'))  # → run (treated as verb)
print(lemma.lemmatize('ran', pos='v'))      # → run
print(lemma.lemmatize('runs', pos='v'))     # → run
```

**When to use Stemming vs Lemmatization:**
- Use **Stemming** when speed matters and the exact meaning is not critical (e.g., search engines, simple classification)
- Use **Lemmatization** when the correct meaning and context of the word is important (e.g., chatbots, translation, sentiment analysis)
- Lemmatization is more accurate but significantly slower

## 9. Named Entity Recognition (NER)

NER is a specialized form of chunking where instead of defining custom rules, the model automatically identifies and groups **named entities** — things like people's names, places, countries, organizations, dates, and monetary values — from a given text.

This is extremely useful in real-world applications. For example:
- In a legal document, extract all **person names** mentioned
- In a news article, extract all **country names** and **dates**
- In a financial report, extract all **monetary values** and **organization names**

NER works by first doing POS tagging and then applying chunking rules specifically designed to detect entities.

**Example:**
```python
sent = "India, officially the Republic of India, is a country in South Asia."
words = nltk.word_tokenize(sent)
pos_tag = nltk.pos_tag(words)
namedEntity = nltk.ne_chunk(pos_tag)
namedEntity.draw()   # shows a graphical tree with entities labeled
```

In the output tree, words like `India`, `Republic of India`, `South Asia` are grouped under the `GPE` (Geo-Political Entity) tag. If there were names of people, they would be grouped under `PERSON`.

**Common NER tags:**
- `PERSON` → names of people
- `GPE` → countries, cities, states
- `ORG` → organizations and companies
- `DATE` → dates and time references
- `MONEY` → monetary values

## 11. Word Vectorization (Word Embedding)

Machine learning models cannot work with raw text — they only understand numbers. **Word Vectorization** is the process of converting words into numbers (vectors) so that machine learning algorithms can process them.

The idea is to represent each word as a point in a multi-dimensional space (a vector). Words that are semantically similar should end up close to each other in this space. For example, the vector for `king` and `queen` should be closer to each other than the vector for `king` and `table`.

Vectorization is done by comparing a given word against the entire **corpus** (collection of all documents). There are several methods for vectorization, ranging from simple frequency-based approaches to complex neural network-based embeddings.

**Most popular methods:**
- Bag of Words (BOW)
- n-gram model
- TF-IDF
- Word2Vec, GloVe (advanced — not covered here)

## 12. Bag of Words (BOW)

The simplest method of vectorization. A **Bag of Words** represents a document as a collection of all the words it contains, keeping track of how many times each word appears (frequency), but completely **ignoring grammar, word order, and sentence structure**. Just like throwing all the words into a bag — you know what's in there, but not the order.

**Example:**

Text1: `"I went to have a cup of coffee but I ended up having lunch with her."`

Text2: `"I don't understand, what is the problem here?"`

**BOW representation:**
```
BOW1 = {I: 2, went: 1, to: 1, have: 1, a: 1, cup: 1, coffee: 1, ended: 1, having: 1, lunch: 1, her: 1}
BOW2 = {I: 1, don't: 1, understand: 1, what: 1, is: 1, problem: 1, here: 1}
```

These dictionaries are then converted into vectors (lists of numbers) which can be fed into ML models.

**Limitation of BOW:** Stop words like `the`, `a`, `is` appear very frequently across all documents. They get high frequency counts, but carry no meaningful information. This misleads the model. That's why TF-IDF was introduced to solve this problem.

**Note:** BOW is a special case of the n-gram model when `n = 1`.

## 13. n-gram Model

BOW ignores word order completely. But in many cases, the order of words matters a lot for meaning. For example, `"not good"` and `"good"` are very different in sentiment, but BOW would treat them similarly. The **n-gram model** fixes this by considering sequences of `n` consecutive words instead of individual words.

- `n = 1` → **Unigram** (same as BOW — individual words)
- `n = 2` → **Bigram** (pairs of consecutive words)
- `n = 3` → **Trigram** (triplets of consecutive words)

**Example:** Text: `"I went to have a cup of coffee"`

- **Unigram:** `[I, went, to, have, a, cup, of, coffee]`
- **Bigram:** `[I went], [went to], [to have], [have a], [a cup], [cup of], [of coffee]`
- **Trigram:** `[I went to], [went to have], [to have a], [have a cup], [a cup of], [cup of coffee]`

**Skip-grams** are a variation where words are not necessarily consecutive — some words in between can be skipped.

Example of **1-skip 2-gram** from `"I don't understand what is the problem":`
```
[I understand], [don't what], [understand is], [what the], [is problem]
```

**Code:**
```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(ngram_range=(2, 2))  # bigrams
vect.fit_transform(["This is an example of n-gram"])
print(vect.get_feature_names())
```

## 14. TF-IDF (Term Frequency – Inverse Document Frequency)

TF-IDF is a smarter method of vectorization that addresses the biggest weakness of BOW — high-frequency common words getting too much importance. TF-IDF assigns a **score to each word in each document** based on two things: how often it appears in that document AND how rare it is across all documents.

The idea is: a word that appears frequently in one document but rarely in others is likely very important to that document. A word that appears in every document (like `the`, `is`, `a`) is not important.

### Term Frequency (TF)
How often does a word appear in a specific document relative to the total words in that document?
$$TF = \frac{\text{Number of times word appears in document}}{\text{Total words in document}}$$

### Inverse Document Frequency (IDF)
How rare is the word across all documents? Rare words get higher IDF scores.
$$IDF = \log_2\left(\frac{\text{Total number of documents}}{\text{Number of documents containing the word}}\right)$$

### TF-IDF Score
$$\text{TF-IDF} = TF \times IDF$$

**Example with 3 documents:**

- Doc1: `"This is an example."`
- Doc2: `"We will see how it works."`
- Doc3: `"IDF can be confusing."`

Since no word appears in more than one document, each word gets a non-zero TF-IDF score. Words that appear in all 3 documents would get IDF = 0, meaning TF-IDF = 0 → they are ignored.

**Note:** sklearn's `TfidfVectorizer` slightly modifies the formula to avoid division by zero and to prevent terms from being completely ignored. It adds `+1` to the IDF formula.

**Code:**
```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

tfid = TfidfVectorizer()
doc = ["This is an example.", "We will see how it works.", "IDF can be confusing"]
doc_vector = tfid.fit_transform(doc)

df = pd.DataFrame(doc_vector.todense(), columns=tfid.get_feature_names())
df   # shows TF-IDF score for each word in each document
```

The resulting table shows each word as a column and each document as a row. The value in each cell is the TF-IDF score — higher means that word is more important for that particular document. Words that appear in all documents get a very low or zero score.

## 15. NLP Pipeline – Full Flow Summary

Here is the typical end-to-end flow when working on any text analytics problem:

```
Raw Text
   ↓
Tokenization  (break into sentences/words)
   ↓
Remove Stop Words & Punctuation  (clean the data)
   ↓
Stemming OR Lemmatization  (reduce to base form)
   ↓
POS Tagging  (label each word's grammar role)
   ↓
Chunking / NER  (group meaningful entities)
   ↓
Vectorization — BOW / n-gram / TF-IDF  (convert to numbers)
   ↓
Machine Learning Model  (classify, predict, recommend)
```

**Key decision points:**
- Speed matters, meaning not critical → **Stemming**
- Meaning and context matter → **Lemmatization**
- Simple frequency-based features → **BOW / Count Vectorizer**
- Word order matters → **n-gram**
- Weighted importance needed → **TF-IDF** (most recommended)
- Extract people/places/organizations → **NER**
- Understand sentence structure → **Parsing**
