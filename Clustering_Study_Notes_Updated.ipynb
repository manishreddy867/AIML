
## 1. What is Clustering?
Clustering is an Unsupervised Machine Learning technique used to group similar data points together.

- Similar points belong to the same cluster.
- Different points belong to different clusters.

Example: Grouping toys by color or shape.



## 2. Goal of Clustering
The goal is to discover hidden patterns and natural groups in data when no target variable is available.



## 3. Applications of Clustering
- Customer Segmentation
- Recommendation Systems
- Fraud Detection
- Image Segmentation
- Market Basket Analysis
- Outlier Detection



## 4. Types of Clustering

### Agglomerative Clustering (Bottom-Up)
- Starts with each point as its own cluster.
- Merges similar clusters step by step.

### Divisive Clustering (Top-Down)
- Starts with one large cluster.
- Splits it into smaller clusters.



## 5. K-Means Clustering

K-Means is one of the most widely used clustering algorithms.

### Steps:
1. Select the value of K.
2. Initialize centroids.
3. Assign points to nearest centroid.
4. Recalculate centroids.
5. Repeat until convergence.



## 6. Centroid

A centroid is the center point of a cluster.



## 7. Inertia

Inertia measures how close data points are to their cluster centroid.

Lower inertia generally indicates better clustering.



## 8. Elbow Method

The Elbow Method helps determine the optimal value of K.

Steps:
- Train K-Means with different K values.
- Calculate inertia for each K.
- Plot K versus Inertia.
- Select the elbow point.



## 9. Advantages of K-Means

- Simple to understand
- Fast execution
- Easy implementation
- Suitable for large datasets



## 10. Limitations of K-Means

- K must be specified beforehand.
- Sensitive to outliers.
- Sensitive to centroid initialization.
- Works best with spherical clusters.



## 11. Hierarchical Clustering

Hierarchical Clustering creates a hierarchy of clusters and is usually visualized using a dendrogram.

Advantages:
- Easy visualization
- No need to specify K initially

Disadvantages:
- Computationally expensive for large datasets



## 12. DBSCAN

DBSCAN stands for Density-Based Spatial Clustering of Applications with Noise.

Important Parameters:
- eps
- min_samples

Advantages:
- Detects outliers
- No need to specify K
- Works with arbitrary cluster shapes



## 13. Cluster Evaluation

### Internal Measures
- Silhouette Score
- Davies-Bouldin Index
- Calinski-Harabasz Score

### External Measures
- Rand Index
- Adjusted Rand Index
